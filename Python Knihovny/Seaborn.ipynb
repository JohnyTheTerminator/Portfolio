```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# import some nice vis settings
sns.set()
# Graphics in the Retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
```

```python
data=r"C:\Users\42073\Desktop\NAFTA3.xlsx"
data = pd.read_excel(data)

```

# BAR PLOT


```python
sns.barplot(x="cesta", y="cena", data=data) 
```

# SCATTER PLOT

```python
sns.scatterplot(x="cesta", y="cena", data=data) 
```

# LINE PLOT 


```python
sns.lineplot(x="cesta", y="cena", data=data)
```

# COUNT PLOT

```python
sns.countplot(x="cena", data=data)
```

# BOX PLOT

```python
sns.boxplot(x="cesta", y="cena", data=data)
```

# PAIR PLOT

```python
sns.pairplot(data=data)
```

# JOINT PLOT

```python
sns.jointplot(x="cesta", y="cena", data=data, kind="kde") 
```

# REG PLOT

```python
sns.regplot(x="cesta", y="cena", data=data) 
```

# KDE PLOT

```python
sns.kdeplot(data=data, x="cesta", y="cena") 
```

# DIS PLOT

```python
sns.displot(data["cena"], kde=False) 
```

# VIOLIN PLOT

```python
sns.violinplot(x="cesta", y="cena", data=data) 

plt.title('VIOLINPLOT')
plt.xlabel('cesta')
plt.ylabel('cena')
          
plt.show()    
```

# SWARM PLOT

```python
sns.swarmplot(x="cesta", y="cena", data=data) 
# Set the title and axis labels
plt.title('SWARMPLOT')
plt.xlabel('cesta')
plt.ylabel('cena')
          
plt.show()          
```
